```python
# import Counter tool
from collections import Counter

# import os module
import os

#import getpass
from getpass import getpass

# file name for saving failed login attempts
file_name = "login_attempts.txt"

# Simple "database" of valid users and their passwords (like a dictionary lookup)
user_db = {"Idris": "123", "Mayowa": "python", "Shomoye": "hello", "Bisong": "secret", "Ade": "pass", "Admin": "admin123"}


# Helper function: Load past failed attempts from file so the system remembers past fails
def load_attempts():
    if os.path.exists(file_name):
        with open(file_name, "r") as file:
            content = file.read().strip()
            if content:
                return [name.strip() for name in content.split(",")]
    return []


# Helper function: Save updated failed attempts to file
def save_attempts(attempt_list):
    with open(file_name, "w") as file:
        file.write(",".join(attempt_list))

# Main login function â€” this does the whole login process
def login():
    
    # Load past failed attempts: i want to use it now
    attempts_list = load_attempts()
    attempts_count = Counter(attempts_list)

    print("=== Simple Login System ===\n")

    username = input("Enter your username: ").strip()

    # Activate admin mode to remove locks 
    if username == "Admin":
        password = getpass("Enter Admin Password")
        if password == user_db["Admin"]:
            print("\nðŸ”‘ ADMIN MODE ACTIVATED!")

    # Find locked accouns for 3 or more fails
            locked_users = [users for users in attempts_count if attempts_count[users] >= 3]
            
            if not locked_users:
                print("No locked accounts.")
            else:
                print("Locked accounts.")
                for user in locked_users:
                    print(f"  - {user} ({attempts_count[user]} failed attempts)")
                
    # Ask admin who to unlock
                unlock_who = input("\nEnter username to unlock (or 'cancel'): ").strip()
                if unlock_who == "cancel":
                    print("No changes made.")
                elif unlock_who in locked_users:
                    # Remove all failed attempts for that user
                    attempts_list = [a for a in attempts_list if a != unlock_who]
                    save_attempts(attempts_list)
                    print(f"ðŸ”“ {unlock_who}'s account has been unlocked!")
                else:
                    print("User not locked or not found.")
            return
        else:
            print("\nâŒ Wrong admin password.")
            return
    
    # Check if account is already locked
    if attempts_count[username] >= 3:
        print("\nðŸ”’ Your account is locked due to too many failed attempts.")
        print("Contact admin to unlock.")
        return
        
        
    # Check if username exists in our database
    if username not in user_db:
        print("\nâŒ Username not found.")
        return

    # Ask for password
    password = getpass(" Enter passwaord: ")

    #correct password?
    if password == user_db[username]:
        print(f"\nâœ… Login successful! Welcome, {username}!")

         # â† NEW: Clear failed attempts on success
        attempts_list = [a for a in attempts_list if a != username]  # Remove all fails for this user
        save_attempts(attempts_list)  # Save the cleaned list
        print("ðŸ”“ Your failed attempts have been cleared!")
        
    else:
        print("\nâŒ Wrong password.")
        
        # Record this failed attempt
        attempts_list.append(username)
        save_attempts(attempts_list)
        
        
    # Update count and show warning
        new_count = attempts_count[username] + 1
        print(f"Failed attempts for {username}: {new_count}/3")

        if new_count >= 3:
            print("ðŸ”’ Your account is now locked!")

# Run the system (call this to start)

login()
    
```
